```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List, Tuple
from sklearn.preprocessing import StandardScaler, RobustScaler, PowerTransformer, QuantileTransformer
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from scipy import stats
import lightgbm as lgb
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import warnings
from datetime import datetime, timedelta
from tqdm import tqdm
import gc
import os

warnings.filterwarnings("ignore")

NORMALIZATION_METHODS = {
    "standard": "Standard Z-score normalization",
    "robust": "Robust scaling using median and IQR",
    "quantile_gaussian": "Quantile transformation to Gaussian distribution",
    "adaptive_power": "Power transformation (Yeo-Johnson)",
    "min_max": "Min-Max normalization to range [0,1]",
    "decimal_scaling": "Decimal scaling based on max value",
    "z_score": "Z-score normalization (mean/std)",
    "median": "Divide by median normalization",
    "sigmoid": "Sigmoid function normalization",
    "tanh_estimator": "Hyperbolic tangent estimator", 
    "mad": "Median Absolute Deviation",
    "rank_gauss": "Rank-based Gaussian transformation"
}

class NormalizationResearch:
    def __init__(
        self,
        cache_path: str = "data",
        start_date: str = "2015-01-01",
        end_date: str = "2022-12-31",
        prediction_horizon: int = 5,
        lookback_window: int = 20,
        market_index: str = "OSEBX.OL",
        fast_mode: bool = False,
        random_seed: int = 42,
        norm_methods: List[str] = None,
        baseline_method: str = "standard"
    ):
        np.random.seed(random_seed)
        torch.manual_seed(random_seed)
        
        if fast_mode:
            print("Fast mode enabled - using shortened time period")
            start_date = (datetime.strptime(end_date, "%Y-%m-%d") - 
                         timedelta(days=365)).strftime("%Y-%m-%d")
        
        self.cache_path = cache_path
        self.start_date = start_date
        self.end_date = end_date
        self.prediction_horizon = prediction_horizon
        self.lookback_window = lookback_window
        self.market_index = market_index
        self.fast_mode = fast_mode
        self.random_seed = random_seed
        
        self.data = None
        self.features = None
        self.feature_names = []
        self.target = None
        self.dates = None
        self.results = {}
        self.baseline_perf = {}
        
        if norm_methods is None:
            self.norm_methods = list(NORMALIZATION_METHODS.keys())
        else:
            self.norm_methods = [method for method in norm_methods if method in NORMALIZATION_METHODS]
            if not self.norm_methods:
                raise ValueError(f"No valid normalization methods specified. Available methods: {list(NORMALIZATION_METHODS.keys())}")
        
        if baseline_method not in NORMALIZATION_METHODS:
            raise ValueError(f"Baseline method '{baseline_method}' not valid. Available methods: {list(NORMALIZATION_METHODS.keys())}")
        
        self.baseline_method = baseline_method
        
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print(f"Using device: {self.device}")
    
    def fetch_data(self) -> pd.DataFrame:
        print(f"Fetching {self.market_index} data from {self.start_date} to {self.end_date}...")
        if not os.path.exists(self.cache_path):
            os.makedirs("data")
        else:
            files = os.listdir(self.cache_path)
            for file_name in files:
                if f"{self.market_index}_{self.start_date}_{self.end_date}.parquet" == file_name:
                    print("Data already downloaded. Loading from file...")
                    self.data = pd.read_parquet(f"={self.cache_path}/{self.start_date}_{self.end_date}")
                    self.data.set_index("Date", inplace=True)
                    return self.data
    
        buffer_start = (datetime.strptime(self.start_date, "%Y-%m-%d") - 
                    timedelta(days=100)).strftime("%Y-%m-%d")
        try:
            market_data = None
            
            try:
                temp_data = yf.download(self.market_index, start=buffer_start, end=self.end_date, progress=False)
                if not temp_data.empty and len(temp_data) > 50:
                    market_data = temp_data
                    print(f"Successfully downloaded data using {self.market_index}")
            except Exception as e:
                print(f"Failed to download with {self.market_index}: {str(e)}")
            
            if market_data is None or market_data.empty:
                raise ValueError(f"Could not download {self.market_index} data")
            
            required_columns = ["Close", "High", "Low", "Open", "Volume"]
            for col in required_columns:
                if col not in market_data.columns:
                    print(f"Warning: {col} not found in data, using Close as fallback")
                    market_data[col] = market_data["Close"] if "Close" in market_data.columns else None
            
            data = pd.DataFrame(index=market_data.index)
            data["close"] = market_data["Close"]
            data["high"] = market_data["High"]
            data["low"] = market_data["Low"]
            data["open"] = market_data["Open"]
            data["volume"] = market_data["Volume"]
            
            for col in ["close", "high", "low", "open"]:
                data[col] = data[col].fillna(method="ffill").fillna(method="bfill")
            
            self.data = data.loc[self.start_date:]
            print(f"Data loaded: {len(self.data)} trading days")

            self.data.to_parquet(f"data/{self.market_index}_{self.start_date}_{self.end_date}")
            
            return self.data
        
        except Exception as e:
            print(f"Error fetching data: {str(e)}")
            raise
    
    def engineer_features(self) -> None:
        if self.data is None:
            raise ValueError("No data available. Call fetch_data() first.")
        
        print("Engineering features...")
        df = self.data.copy()
        
        if len(df) < 30:
            print(f"Warning: Only {len(df)} data points available. Need at least 30 for reliable features.")
            if len(df) < 10:
                raise ValueError(f"Insufficient data points ({len(df)}) for feature engineering")
        
        try:
            for period in [1, 2, 3, 5, 10, 21]:
                if period < len(df):
                    df[f"return_{period}d"] = df["close"].pct_change(period)
            
            for period in [10, 20, 50, 100]:
                if period < len(df):
                    df[f"ma_{period}d"] = df["close"].rolling(window=period).mean()
                    df[f"ma_dist_{period}d"] = (df["close"] / df[f"ma_{period}d"] - 1) * 100
            
            if "return_1d" in df.columns:
                for period in [10, 21]:
                    if period < len(df):
                        df[f"vol_{period}d"] = df["return_1d"].rolling(window=period).std() * np.sqrt(252)
            
            if "ma_20d" in df.columns and "ma_50d" in df.columns:
                df["ma_cross_20_50"] = (df["ma_20d"] > df["ma_50d"]).astype(int)
            
            df["high_low_ratio"] = df["high"] / df["low"]
            df["close_open_ratio"] = df["close"] / df["open"]
            
            if "volume" in df.columns and df["volume"].sum() > 0:
                df["volume_ma_10d"] = df["volume"].rolling(window=10).mean()
                df["volume_ratio"] = df["volume"] / df["volume_ma_10d"]
            
            if self.prediction_horizon < len(df):
                df[f"target_return_{self.prediction_horizon}d"] = df["close"].pct_change(
                    self.prediction_horizon
                ).shift(-self.prediction_horizon)
                
                df["target"] = (df[f"target_return_{self.prediction_horizon}d"] > 0).astype(int)
            else:
                print(f"Warning: Not enough data points for prediction horizon of {self.prediction_horizon}")
                shorter_horizon = max(1, min(3, len(df) // 3))
                print(f"Using shorter prediction horizon: {shorter_horizon} days")
                df[f"target_return_{shorter_horizon}d"] = df["close"].pct_change(
                    shorter_horizon
                ).shift(-shorter_horizon)
                
                df["target"] = (df[f"target_return_{shorter_horizon}d"] > 0).astype(int)
            
            self.feature_names = [col for col in df.columns 
                                if not col.startswith("target") 
                                and col != "close"
                                and col != "high"
                                and col != "low"
                                and col != "open"
                                and col != "volume"]
                                
            if not self.feature_names:
                print("Warning: No features generated. Creating basic features.")
                df["log_return"] = np.log(df["close"] / df["close"].shift(1))
                self.feature_names = ["log_return"]
            
            df = df.dropna()
            
            self.features = df[self.feature_names]
            self.target = df["target"]
            self.dates = df.index
            self.actual_returns = df[f"target_return_{self.prediction_horizon}d"] if f"target_return_{self.prediction_horizon}d" in df.columns else df["target_return_3d"]
            
            print(f"Features prepared: {len(self.features)} samples with {len(self.feature_names)} features")
            
            if self.fast_mode and len(self.feature_names) > 15:
                print(f"Fast mode: limiting to 15 features (from {len(self.feature_names)})")
                self.feature_names = self.feature_names[:15]
                self.features = self.features[self.feature_names]
                
        except Exception as e:
            print(f"Error in feature engineering: {str(e)}")
            import traceback
            traceback.print_exc()
            raise
    
    def train_test_split(self, test_size: float = 0.2) -> Tuple:
        if self.features is None or self.target is None:
            raise ValueError("Features not engineered. Call engineer_features() first.")
        
        split_idx = int(len(self.features) * (1 - test_size))
        
        X_train = self.features.iloc[:split_idx]
        X_test = self.features.iloc[split_idx:]
        y_train = self.target.iloc[:split_idx]
        y_test = self.target.iloc[split_idx:]
        dates_test = self.dates[split_idx:]
        returns_test = self.actual_returns.iloc[split_idx:]
        
        return X_train, X_test, y_train, y_test, dates_test, returns_test
    
    def apply_normalization(self, method: str, X_train: pd.DataFrame, X_test: pd.DataFrame) -> Tuple[np.ndarray, np.ndarray]:
        if method == "standard":
            scaler = StandardScaler()
            X_train_norm = scaler.fit_transform(X_train)
            X_test_norm = scaler.transform(X_test)
            
        elif method == "robust":
            scaler = RobustScaler()
            X_train_norm = scaler.fit_transform(X_train)
            X_test_norm = scaler.transform(X_test)
            
        elif method == "quantile_gaussian":
            scaler = QuantileTransformer(
                n_quantiles=1000, 
                output_distribution="normal",
                random_state=self.random_seed
            )
            X_train_norm = scaler.fit_transform(X_train)
            X_test_norm = scaler.transform(X_test)
            
        elif method == "adaptive_power":
            scaler = PowerTransformer(method="yeo-johnson", standardize=True)
            X_train_norm = scaler.fit_transform(X_train)
            X_test_norm = scaler.transform(X_test)
            
        elif method == "min_max":
            X_train_norm = X_train.copy().values
            X_test_norm = X_test.copy().values
            
            for col in range(X_train.shape[1]):
                col_min = X_train.iloc[:, col].min()
                col_max = X_train.iloc[:, col].max()
                col_range = col_max - col_min
                
                if col_range != 0:
                    X_train_norm[:, col] = (X_train.iloc[:, col].values - col_min) / col_range
                    X_test_norm[:, col] = (X_test.iloc[:, col].values - col_min) / col_range
                    X_test_norm[:, col] = np.clip(X_test_norm[:, col], 0, 1)
        
        elif method == "decimal_scaling":
            X_train_norm = X_train.copy().values
            X_test_norm = X_test.copy().values
            
            for col in range(X_train.shape[1]):
                max_abs_val = np.max(np.abs(X_train.iloc[:, col].values))
                j = 0
                while max_abs_val >= 1:
                    max_abs_val /= 10
                    j += 1
                
                if j > 0:
                    divisor = 10 ** j
                    X_train_norm[:, col] = X_train.iloc[:, col].values / divisor
                    X_test_norm[:, col] = X_test.iloc[:, col].values / divisor
        
        elif method == "z_score":
            X_train_norm = X_train.copy().values
            X_test_norm = X_test.copy().values
            
            for col in range(X_train.shape[1]):
                col_mean = X_train.iloc[:, col].mean()
                col_std = X_train.iloc[:, col].std()
                
                if col_std != 0:
                    X_train_norm[:, col] = (X_train.iloc[:, col].values - col_mean) / col_std
                    X_test_norm[:, col] = (X_test.iloc[:, col].values - col_mean) / col_std
        
        elif method == "median":
            X_train_norm = X_train.copy().values
            X_test_norm = X_test.copy().values
            
            for col in range(X_train.shape[1]):
                col_median = X_train.iloc[:, col].median()
                
                if col_median != 0:
                    X_train_norm[:, col] = X_train.iloc[:, col].values / col_median
                    X_test_norm[:, col] = X_test.iloc[:, col].values / col_median
        
        elif method == "sigmoid":
            X_train_norm = X_train.copy().values
            X_test_norm = X_test.copy().values
            
            for col in range(X_train.shape[1]):
                X_train_norm[:, col] = 1 / (1 + np.exp(-X_train.iloc[:, col].values))
                X_test_norm[:, col] = 1 / (1 + np.exp(-X_test.iloc[:, col].values))
        
        elif method == "tanh_estimator":
            X_train_norm = X_train.copy().values
            X_test_norm = X_test.copy().values
            
            for col in range(X_train.shape[1]):
                col_mean = X_train.iloc[:, col].mean()
                col_std = X_train.iloc[:, col].std()
                
                if col_std != 0:
                    X_train_norm[:, col] = 0.5 * (np.tanh(0.01 * ((X_train.iloc[:, col].values - col_mean) / col_std)) + 1)
                    X_test_norm[:, col] = 0.5 * (np.tanh(0.01 * ((X_test.iloc[:, col].values - col_mean) / col_std)) + 1)
        
        elif method == "mad":
            X_train_norm = X_train.copy().values
            X_test_norm = X_test.copy().values
            
            for col in range(X_train.shape[1]):
                col_median = X_train.iloc[:, col].median()
                col_mad = np.median(np.abs(X_train.iloc[:, col].values - col_median))
                
                if col_mad != 0:
                    X_train_norm[:, col] = (X_train.iloc[:, col].values - col_median) / col_mad
                    X_test_norm[:, col] = (X_test.iloc[:, col].values - col_median) / col_mad
        
        elif method == "robust_zscore":
            X_train_norm = X_train.values.copy()
            X_test_norm = X_test.values.copy()
            
            for col in range(X_train.shape[1]):
                col_data = X_train.iloc[:, col].values
                
                lower_bound = np.percentile(col_data, 5)
                upper_bound = np.percentile(col_data, 95)
                col_data_winsorized = np.clip(col_data, lower_bound, upper_bound)
                
                median = np.median(col_data_winsorized)
                mad = np.median(np.abs(col_data_winsorized - median))
                mad = max(mad, 1e-8)
                
                X_train_norm[:, col] = (X_train.iloc[:, col].values - median) / (1.4826 * mad)
                X_test_norm[:, col] = (X_test.iloc[:, col].values - median) / (1.4826 * mad)
        
        elif method == "rank_gauss":
            X_train_norm = np.zeros_like(X_train.values)
            X_test_norm = np.zeros_like(X_test.values)
            
            for col in range(X_train.shape[1]):
                train_col = X_train.iloc[:, col].values.copy()
                ranks = stats.rankdata(train_col, method='average') - 1
                ranks = ranks / (len(train_col) - 1)
                X_train_norm[:, col] = stats.norm.ppf(ranks)
                
                for i, val in enumerate(X_test.iloc[:, col].values):
                    rank = np.sum(train_col < val) / len(train_col)
                    if rank <= 0:
                        rank = 0.5 / len(train_col)
                    elif rank >= 1:
                        rank = 1 - 0.5 / len(train_col)
                    X_test_norm[i, col] = stats.norm.ppf(rank)
            
            X_train_norm = np.clip(X_train_norm, -10, 10)
            X_test_norm = np.clip(X_test_norm, -10, 10)
            X_train_norm = np.nan_to_num(X_train_norm)
            X_test_norm = np.nan_to_num(X_test_norm)
        
        else:
            raise ValueError(f"Unknown normalization method: {method}")
        
        return X_train_norm, X_test_norm
    
    def prepare_lstm_data(self, X_train_norm: np.ndarray, X_test_norm: np.ndarray, 
                         y_train: np.ndarray, y_test: np.ndarray) -> Tuple:
        lookback = self.lookback_window
        
        X_train_seq = []
        y_train_seq = []
        
        for i in range(lookback, len(X_train_norm)):
            X_train_seq.append(X_train_norm[i-lookback:i])
            y_train_seq.append(y_train.iloc[i])
        
        X_test_seq = []
        y_test_seq = []
        
        for i in range(lookback, len(X_test_norm)):
            X_test_seq.append(X_test_norm[i-lookback:i])
            y_test_seq.append(y_test.iloc[i])
        
        X_train_seq = np.array(X_train_seq)
        y_train_seq = np.array(y_train_seq)
        X_test_seq = np.array(X_test_seq)
        y_test_seq = np.array(y_test_seq)
        
        X_train_tensor = torch.FloatTensor(X_train_seq)
        y_train_tensor = torch.FloatTensor(y_train_seq).unsqueeze(1)
        
        train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
        
        return train_loader, torch.FloatTensor(X_test_seq), torch.FloatTensor(y_test_seq)

    class BidirectionalLSTM(nn.Module):
        def __init__(self, input_dim: int, hidden_dim: int = 64):
            super().__init__()
            
            self.lstm = nn.LSTM(
                input_dim, 
                hidden_dim, 
                batch_first=True, 
                bidirectional=True
            )
            
            self.fc = nn.Sequential(
                nn.Linear(hidden_dim * 2, 32),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(32, 1),
                nn.Sigmoid()
            )
            
        def forward(self, x: torch.Tensor) -> torch.Tensor:
            lstm_out, _ = self.lstm(x)
            lstm_out = lstm_out[:, -1]
            return self.fc(lstm_out)
    
    def train_lightgbm(self, X_train: np.ndarray, y_train: np.ndarray) -> lgb.LGBMClassifier:
        print("Training LightGBM model...")
        
        params = {
            'learning_rate': 0.05,
            'n_estimators': 500 if self.fast_mode else 1000,
            'max_depth': 8,
            'num_leaves': 77,
            'subsample': 0.8,
            'colsample_bytree': 0.8,
            'min_child_samples': 20,
            'random_state': self.random_seed,
            'class_weight': 'balanced',
            'min_gain_to_split': 0.0,
            'min_child_weight': 0.001,
            'n_jobs': -1,
            # 'verbose': -1  # Suppress warnings
        }
        
        model = lgb.LGBMClassifier(**params)
        model.fit(X_train, y_train)
        
        return model
    
    def train_lstm(self, train_loader: DataLoader, input_dim: int) -> nn.Module:
        print("Training LSTM model...")
        
        model = self.BidirectionalLSTM(input_dim, hidden_dim=64).to(self.device)
        criterion = nn.BCELoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)
        
        num_epochs = 3 if self.fast_mode else 35
        
        for epoch in range(num_epochs):
            model.train()
            total_loss = 0
            
            for inputs, labels in train_loader:
                inputs = inputs.to(self.device)
                labels = labels.to(self.device)
                
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                loss.backward()
                optimizer.step()
                
                total_loss += loss.item()
            
            avg_loss = total_loss / len(train_loader)
            print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")
        
        return model
    
    def evaluate_models(self, lgb_model: lgb.LGBMClassifier, lstm_model: nn.Module, 
                      X_test: np.ndarray, y_test: np.ndarray, 
                      X_test_lstm: torch.Tensor, y_test_lstm: torch.Tensor) -> Dict:
        print("Evaluating models...")
        
        lgb_preds = lgb_model.predict(X_test)
        lgb_preds_proba = lgb_model.predict_proba(X_test)[:, 1]
        
        lstm_model.eval()
        with torch.no_grad():
            lstm_outputs = lstm_model(X_test_lstm.to(self.device))
            lstm_preds_proba = lstm_outputs.cpu().numpy().flatten()
            lstm_preds = (lstm_preds_proba > 0.5).astype(int)
        
        metrics = {}
        
        for model_name, preds, y_true in [
            ("lightgbm", lgb_preds, y_test), 
            ("lstm", lstm_preds, y_test_lstm.numpy())
        ]:
            metrics[model_name] = {
                "accuracy": accuracy_score(y_true, preds),
                "f1": f1_score(y_true, preds),
                "precision": precision_score(y_true, preds),
                "recall": recall_score(y_true, preds)
            }
            
            print(f"{model_name.upper()} - Accuracy: {metrics[model_name]['accuracy']:.4f}, "
                 f"F1: {metrics[model_name]['f1']:.4f}")
        
        return metrics
    
    def run_normalization_comparison(self) -> Dict:
        print("\n" + "="*50)
        print("NORMALIZATION TECHNIQUES COMPARISON")
        print("="*50)
        
        try:
            self.fetch_data()
            self.engineer_features()
            
            if self.features is None or len(self.features) < 30:
                print("ERROR: Not enough data for analysis after preprocessing")
                print(f"Available samples: {0 if self.features is None else len(self.features)}")
                return {}
                
            X_train, X_test, y_train, y_test, dates_test, returns_test = self.train_test_split()
            
            methods_to_evaluate = self.norm_methods.copy()
            if self.baseline_method not in methods_to_evaluate:
                methods_to_evaluate.append(self.baseline_method)
            
            results = {method: {} for method in methods_to_evaluate}
            
            baseline_perf = {}
            for method in tqdm(methods_to_evaluate, desc="Evaluating normalization methods"):
                print(f"\nTesting normalization method: {method}")
                
                try:
                    X_train_norm, X_test_norm = self.apply_normalization(method, X_train, X_test)
                    
                    lgb_model = self.train_lightgbm(X_train_norm, y_train)
                    
                    train_loader, X_test_lstm, y_test_lstm = self.prepare_lstm_data(
                        X_train_norm, X_test_norm, y_train, y_test
                    )
                    
                    lstm_model = self.train_lstm(train_loader, X_train.shape[1])
                    
                    model_metrics = self.evaluate_models(
                        lgb_model, lstm_model, 
                        X_test_norm, y_test, 
                        X_test_lstm, y_test_lstm
                    )
                    
                    results[method] = model_metrics
                    
                    if method == self.baseline_method:
                        self.baseline_perf = model_metrics.copy()
                    
                except Exception as e:
                    print(f"Error evaluating {method}: {str(e)}")
                    results[method] = {"error": str(e)}
                
                if 'lgb_model' in locals():
                    del lgb_model
                if 'lstm_model' in locals():
                    del lstm_model
                if 'train_loader' in locals():
                    del train_loader
                if 'X_test_lstm' in locals():
                    del X_test_lstm
                if 'y_test_lstm' in locals():
                    del y_test_lstm
                gc.collect()
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
            
            self.results = results
            return results
            
        except Exception as e:
            print(f"Error in normalization comparison: {str(e)}")
            import traceback
            traceback.print_exc()
            return {}
    
    def plot_results(self) -> None:
        if not self.results:
            print("No results to plot. Run run_normalization_comparison() first.")
            return
        
        valid_results = {method: results for method, results in self.results.items() 
                       if isinstance(results, dict) and "error" not in results}
        
        if not valid_results:
            print("No valid results to plot. All methods encountered errors.")
            return
        
        data = []
        baseline_data = []
        for norm_method, model_results in valid_results.items():
            for model_name, metrics in model_results.items():
                for metric_name, value in metrics.items():
                    data_point = {
                        "Normalization": norm_method,
                        "Model": model_name,
                        "Metric": metric_name,
                        "Value": value
                    }
                    
                    data.append(data_point)
                    
                    if norm_method == self.baseline_method:
                        baseline_data.append(data_point)
        
        df_results = pd.DataFrame(data)
        df_baseline = pd.DataFrame(baseline_data) if baseline_data else None
        
        plt.figure(figsize=(15, 10))
        
        for i, metric in enumerate(["accuracy", "f1"]):
            plt.subplot(2, 1, i+1)
            df_metric = df_results[df_results["Metric"] == metric]
            
            plot = sns.barplot(x="Normalization", y="Value", hue="Model", data=df_metric)
            
            if df_baseline is not None:
                baseline_metric = df_baseline[df_baseline["Metric"] == metric]
                if not baseline_metric.empty:
                    for model in ["lightgbm", "lstm"]:
                        baseline_val = baseline_metric[baseline_metric["Model"] == model]["Value"].values
                        if len(baseline_val) > 0:
                            plt.axhline(
                                y=baseline_val[0], 
                                linestyle='--', 
                                color='red' if model == "lightgbm" else 'blue',
                                alpha=0.7,
                                label=f"{self.baseline_method} {model} baseline"
                            )
            
            plt.title(f"{metric.capitalize()} by Normalization Method")
            plt.xlabel("Normalization Method")
            plt.ylabel(metric.capitalize())
            plt.grid(axis="y", linestyle="--", alpha=0.7)
            plt.legend(title="Model")
            
            for container in plt.gca().containers:
                plot.bar_label(container, fmt="%.3f")
        
        plt.tight_layout()
        plt.show()
        
        plt.figure(figsize=(15, 10))
        
        for i, model_name in enumerate(["lightgbm", "lstm"]):
            plt.subplot(2, 1, i+1)
            
            df_model = df_results[df_results["Model"] == model_name]
            
            pivot = df_model.pivot(index="Normalization", columns="Metric", values="Value")
            
            sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt=".3f", cbar_kws={"label": "Value"})
            
            plt.title(f"{model_name.upper()} Performance Across Normalization Methods")
            plt.tight_layout()
        
        plt.tight_layout()
        plt.show()
        
        model_ranks = {}
        best_methods = {}
        
        for model_name in ["lightgbm", "lstm"]:
            rank_data = []
            for metric in ["accuracy", "f1"]:
                df_subset = df_results[(df_results["Model"] == model_name) & 
                                      (df_results["Metric"] == metric)]
                
                if df_subset.empty:
                    continue
                    
                df_subset = df_subset.copy()
                df_subset["Rank"] = df_subset["Value"].rank(ascending=False)
                
                for _, row in df_subset.iterrows():
                    rank_data.append({
                        "Normalization": row["Normalization"],
                        "Metric": metric,
                        "Rank": row["Rank"],
                        "Value": row["Value"],
                        "IsBaseline": row["Normalization"] == self.baseline_method
                    })
            
            if rank_data:
                df_ranks = pd.DataFrame(rank_data)
                model_ranks[model_name] = df_ranks
                
                avg_ranks = df_ranks.groupby("Normalization").agg({
                    "Rank": "mean",
                    "IsBaseline": "first"
                }).reset_index()
                avg_ranks = avg_ranks.sort_values("Rank")
                
                best_method = avg_ranks.iloc[0]["Normalization"]
                best_methods[model_name] = best_method
        
        for model_name, df_ranks in model_ranks.items():
            plt.figure(figsize=(10, 6))
            
            avg_ranks = df_ranks.groupby("Normalization").agg({
                "Rank": "mean",
                "IsBaseline": "first"
            }).reset_index()
            avg_ranks = avg_ranks.sort_values("Rank")
            
            bars = plt.bar(
                avg_ranks["Normalization"], 
                avg_ranks["Rank"],
                color=[("red" if row["IsBaseline"] else 
                       ("green" if row["Normalization"] == best_methods[model_name] else "skyblue")) 
                       for _, row in avg_ranks.iterrows()]
            )
            
            for bar in bars:
                height = bar.get_height()
                plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                        f"{height:.2f}", ha="center", va="bottom")
            
            plt.title(f"Average Rank of Normalization Methods for {model_name.upper()} (Lower is Better)")
            plt.xlabel("Normalization Method")
            plt.ylabel("Average Rank")
            plt.grid(axis="y", linestyle="--", alpha=0.5)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        
        rank_data = []
        for model_name in ["lightgbm", "lstm"]:
            for metric in ["accuracy", "f1"]:
                df_subset = df_results[(df_results["Model"] == model_name) & 
                                      (df_results["Metric"] == metric)]
                
                if df_subset.empty:
                    continue
                    
                df_subset = df_subset.copy()
                df_subset["Rank"] = df_subset["Value"].rank(ascending=False)
                
                for _, row in df_subset.iterrows():
                    rank_data.append({
                        "Normalization": row["Normalization"],
                        "Model": model_name,
                        "Metric": metric,
                        "Rank": row["Rank"],
                        "IsBaseline": row["Normalization"] == self.baseline_method
                    })
        
        if not rank_data:
            print("No ranking data available.")
            return
            
        df_ranks = pd.DataFrame(rank_data)
        
        avg_ranks = df_ranks.groupby("Normalization").agg({
            "Rank": "mean",
            "IsBaseline": "first"
        }).reset_index()
        avg_ranks = avg_ranks.sort_values("Rank")
        
        plt.figure(figsize=(10, 6))
        bars = plt.bar(
            avg_ranks["Normalization"], 
            avg_ranks["Rank"],
            color=[("red" if row["IsBaseline"] else "skyblue") for _, row in avg_ranks.iterrows()]
        )
        
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f"{height:.2f}", ha="center", va="bottom")
        
        plt.title("Overall Average Rank of Normalization Methods (Lower is Better)")
        plt.xlabel("Normalization Method")
        plt.ylabel("Average Rank")
        plt.grid(axis="y", linestyle="--", alpha=0.5)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
        
        overall_best_method = avg_ranks.iloc[0]["Normalization"]
        print("\n" + "="*50)
        print("NORMALIZATION METHOD PERFORMANCE SUMMARY")
        print("="*50)
        
        print(f"\nOverall best normalization method: {overall_best_method}")
        
        print("\nBest normalization method per model:")
        for model, method in best_methods.items():
            print(f"  {model.upper()}: {method}")
            
            if method in self.results and model in self.results[method]:
                metrics = self.results[method][model]
                print(f"    Metrics: Accuracy={metrics['accuracy']:.4f}, F1={metrics['f1']:.4f}")
        
        print("\nBaseline performance:")
        for model, metrics in self.baseline_perf.items():
            if isinstance(metrics, dict):
                print(f"  {model.upper()}: Accuracy={metrics['accuracy']:.4f}, F1={metrics['f1']:.4f}")
        
        print("\nPerformance improvement over baseline:")
        for model_name, best_method in best_methods.items():
            if (best_method in self.results and 
                model_name in self.results[best_method] and 
                model_name in self.baseline_perf):
                
                best_metrics = self.results[best_method][model_name]
                base_metrics = self.baseline_perf[model_name]
                
                print(f"\n  {model_name.upper()} - Best Method: {best_method} vs Baseline: {self.baseline_method}")
                
                for metric in ["accuracy", "f1", "precision", "recall"]:
                    if metric in best_metrics and metric in base_metrics:
                        best_val = best_metrics[metric]
                        base_val = base_metrics[metric]
                        diff = best_val - base_val
                        pct_change = (diff / base_val) * 100 if base_val != 0 else float('inf')
                        print(f"    {metric}: {best_val:.4f} vs {base_val:.4f} ({pct_change:+.2f}%)")

def main():
    print("\n" + "="*50)
    print("Impact of Normalization Techniques on ML Models for Stock Prediction")
    print("="*50)
    
    norm_methods = [
        "min_max",
        "decimal_scaling",
        "z_score",
        "median",
        "sigmoid",
        "tanh_estimator",
        "rank_gauss"
    ]
    
    baseline = "min_max"
    
    fast_mode = False
    
    research = NormalizationResearch(
        cache_path="data",
        fast_mode=fast_mode,
        start_date="2015-01-01",
        end_date="2022-12-31",
        norm_methods=norm_methods,
        baseline_method=baseline,
        market_index="OSEBX.OL"  # Oslo Stock Exchange
    )
    
    research.run_normalization_comparison()
    research.plot_results()

if __name__ == "__main__":
    main()
```
